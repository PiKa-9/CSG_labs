```python
from PIL import Image, ImageDraw
```

```python
img = Image.new("RGB", size=(960, 540), color="#FFFFFF")
draw = ImageDraw.Draw(img)
```

```python
with open('../data/DS8.txt', 'r') as f:
    for line in f:
        # x and y are swapped in the data file
        y, x = tuple(map(int, line.split()))
        draw.point((x, y), fill="#B9AB8E")
```
